```python
#Bước 1 Import thư viện turtle
import turtle

# Bước 2 tạo đối tượng
t=turtle.Turtle()
t.shape('turtle')
t.color('blue')
t.pensize(2)
t.hideturtle()


#Bước 3 vẽ



t.fillcolor('blue')
t.begin_fill()

t.forward(80)
t.left(90)
t.forward(200)
t.left(90)
t.forward(80)
t.left(90)
t.forward(200)
t.end_fill()

t.penup()
t.left(90)
t.forward(160)
t.pendown()

t.fillcolor('white')
t.begin_fill()
t.left(90)
t.forward(200)
t.left(90)
t.forward(80)
t.left(90)
t.forward(200)
t.left(90)
t.forward(80)
t.end_fill()

t.penup()
t.forward(80)
t.pendown()

t.fillcolor('red')
t.begin_fill()
t.left(90)
t.forward(200)
t.left(90)
t.forward(80)
t.left(90)
t.forward(200)
t.left(90)
t.forward(80)
t.end_fill()




#Bước 4 hoàn thành
turtle.done()
```
